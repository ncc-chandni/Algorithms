```python
import numpy as np
x = np.array([[1],[2],[3],[4],[5]])
y = np.array([[2],[4],[5],[4],[5]])
```

```python
x = np.concatenate([np.ones((len(x),1)), x], axis = 1)
```

```python
theta_init = np.zeros((2,1))
learning_rate = 0.01
num_iterations = 1000
```

```python
def cost_function(x, y, theta):
    m = len(y)
    yhat = x.dot(theta)
    J = np.sum((yhat - y)**2)/(2*m)
    return J
```

```python
def gradient_descent(x, y, theta):
    m = len(y)
    yhat = x.dot(theta)
    grad = (1/m)*x.T.dot(yhat - y)
    return grad
```

```python
def stochastic_gradient_descent(x, y, theta_init, learning_rate, num_iterations):
    theta = theta_init
    m = len(y)
    J_history = np.zeros(num_iterations)

    for i in range(num_iterations):
        rand_index = np.random.randint(0,m)
        x_i = np.array([x[rand_index,:]]).reshape(1,-1)
        y_i = np.array([y[rand_index,:]]).reshape(1,-1)
        grad = gradient_descent(x_i, y_i, theta)
        theta = theta - learning_rate*grad
        J_history[i] = cost_function(x, y, theta)
    return theta, J_history
```

```python
theta, J_history = stochastic_gradient_descent(x, y, theta_init, learning_rate, num_iterations)
```

```python
import matplotlib.pyplot as plt
plt.plot(J_history)
plt.xlabel("No of iterations")
plt.ylabel("Cost")
plt.show()
```

```python
print(f"Optimized weight: {theta[1]}")
print(f"Optimized bias: {theta[0]}")
print(f"Minimum cost values: {J_history[num_iterations-1]}")
```
