```python
import numpy as np
```

```python
x = np.random.randn(10,1)
y = 5 * x + np.random.rand()

m = 0.0
b = 0.0

learning_rate = 0.01
```

```python
def gradient_descent(x, y, m, b, learning_rate):
  dldm = 0
  dldb = 0
  N = x.shape[0]
  for xi, yi in zip(x,y):
    dldm += -2 * xi * (yi - (m * xi + b))
    dldb += -2 * (yi - (m * xi + b))

  m = m - learning_rate * (1/N) * dldm
  b = b - learning_rate * (1/N) * dldb

  return m , b
```

```python
for epoch in range(400):
  m, b = gradient_descent(x, y, m, b, learning_rate)
  yhat = m * x + b
  loss = np.divide(np.sum((y - yhat)**2, axis = 0), x.shape[0])
  print(f"Epochs:{epoch}, Loss: {loss}, m: {m}, b: {b}")
```
